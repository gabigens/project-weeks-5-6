```python
# Merging 2 tables together on client_id so to enreach the Foot print tables with information with client was in wich groop (control or test)
new_table = pd.merge(digital_fprint, experiment_roster, on = 'client_id', how = 'inner')
```

```python
new_table
```

```python
# Spliting for 2 tables 2 groops Test groop and Control groop

var_tables = {}
for var in new_table['Variation'].unique():
    var_tables[var] = new_table[new_table['Variation'] ==var]


test_table = var_tables['Test']
control_table = var_tables['Control']
```

```python
# 1.KPI # Completion Rate - how many users reached the final step - CONFIRM The proportion of users who reach the final 'confirm' step.
```

```python
# For Contral groop:
# 1. calculate the nulber of clients with 'start' process step
# 2. calculate number of clients with 'confirm' process step
# 3. Divide number of clients with CONFIRM status on clienys with START status - Completion Rate

total_client_start_c = control_table[control_table['process_step']== 'start']['visit_id'].nunique()
total_client_confirm_c = control_table[control_table['process_step']== 'confirm']['visit_id'].nunique()

completion_rate_control = total_client_confirm_c/total_client_start_c

completion_rate_control = completion_rate_control

completion_rate_control
```

```python
# For Test groop:

total_client_start_t = test_table[test_table['process_step']== 'start']['visit_id'].nunique()
total_client_confirm_t = test_table[test_table['process_step']== 'confirm']['visit_id'].nunique()

completion_rate_test = total_client_confirm_t/total_client_start_t

completion_rate_test = completion_rate_test

completion_rate_test
```

```python
# Map Process Steps to Numbers

step_map = {'start': 0, 'step_1': 1, 'step_2': 2, 'step_3': 3, 'confirm': 4}

test_table['step_number'] = test_table['process_step'].map(step_map)
```

```python
# Group data by visit_id to isolate sequences of steps for each visit
# Use the diff() method on the numerical steps to identify negative differences, which indicate a backward step
```

```python
test_table
```

```python
# TEST GROUP
test_table['step_diff'] = test_table.groupby('visit_id')['step_number'].diff()
test_table['is_error'] = test_table['step_diff'].apply(lambda x: True if x < 0 else False)

error_count_per_visitid_test = test_table.groupby('visit_id')['is_error'].sum()
total_steps_per_visit_test = test_table.groupby('visit_id')['process_step'].count()

error_rate_test = error_count_per_visitid_test/total_steps_per_visit_test

error_rate_test
```

```python
average_test_error_rate = error_rate_test.mean()
average_test_error_rate 
```

```python
# CONTROL GROUP

control_table['step_number'] = control_table['process_step'].map(step_map)

control_table['step_diff'] = control_table.groupby('visit_id')['step_number'].diff()
control_table['is_error'] = control_table['step_diff'].apply(lambda x: True if x < 0 else False)

error_count_per_visitid_control = control_table.groupby('visit_id')['is_error'].sum()
total_steps_per_visit_control = control_table.groupby('visit_id')['process_step'].count()

error_rate_control = error_count_per_visitid_control/total_steps_per_visit_control
error_rate_control
```

```python
import scipy.stats as st
from scipy.stats import chi2_contingency
```

```python
#use the chi-square test for this scenario to see if there's a significant difference in completion rates between your control and test groups.
#The chi-square test is appropriate here because it helps determine if the distribution of success and failure is independent of the group (control vs. test)
```

```python
contingency_table = [
    [total_client_confirm_c, total_client_start_c - total_client_confirm_c],  # Control group
    [total_client_confirm_t, total_client_start_t - total_client_confirm_t]   # Test group
]

# Performing the chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Results
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
```

```python
#HO: The error rate of the control group equals the completion rate of the test group 
#H1: The error rate of the control group is not equal to the completion rate of the test group 
```

```python
total_error_test = error_count_per_visitid_test.sum()
total_step_test = total_steps_per_visit_test.sum()

total_error_control = error_count_per_visitid_control.sum()
total_step_control = total_steps_per_visit_control.sum()

# Contingency table
contingency_table = [
    [total_error_control, total_step_control - total_error_control],  # Control group
    [total_error_test, total_step_test - total_error_test]            # Test group
]

# Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Results
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
#H0: The completion rate of the control group equals the completion rate of the test group ((p_c = p_t)).
#H1: The completion rate of the control group is not equal to the completion rate of the test group ((p_c \neq p_t)).
```

```python
# Number of successes and trials
success = [total_client_confirm_c, total_client_confirm_t]
trials = [total_client_start_c, total_client_start_t]

# Conduct the two-proportion z-test
stat, p_value = proportions_ztest(success, trials)

# Output the test results
print(f"Z-statistic: {stat}")
print(f"P-value: {p_value}")

```
